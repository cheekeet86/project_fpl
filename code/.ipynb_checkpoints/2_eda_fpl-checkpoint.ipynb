### Explore df_fpl

```python
print(df_fpl['full'].nunique())

df_fpl['full'].unique()
```

```python
cols = ['full','team']

df_fpl_sum = df_fpl.groupby(cols).sum()
df_fpl_sum.reset_index(inplace=True)
df_fpl_sum
```

```python
cond = df_fpl_sum['total_points'] > 0

df_fpl_sum = df_fpl_sum[cond]
df_fpl_sum
```

```python
cols = ['team','total_points']
order = [True,False]

df_fpl_sum.sort_values(by=cols,ascending=order,inplace=True)
```

```python
df_fpl_sum['team_rank'] = 0
```

```python
df_fpl_sum_2 = df_fpl_sum[0:0]
```

```python
team_list = df_fpl_sum['team'].unique()
team_list
```

```python
for team in team_list:
    
    cond = df_fpl_sum['team'] == team
    df_fpl_team = df_fpl_sum[cond]
    
    df_fpl_team['team_rank'] = df_fpl_team['total_points'].rank(method='min',ascending=False)
    
    #cols = ['full','team','total_points','team_rank']
    #print(df_fpl_team[cols])
    
    df_fpl_sum_2 = df_fpl_sum_2.append(df_fpl_team,ignore_index=True)
```

```python
cols = ['full','team','total_points']
df_fpl_totalpts = df_fpl_sum_2[cols]
df_fpl_totalpts.sort_values(by='total_points',ascending=False)
```

```python
cond = df_fpl_sum_2['team_rank'] <= 3

# drop 1 record (joint 3rd position)
cond_2 = df_fpl_sum_2['full'] != 'George Baldock'

df_fpl_top3 = df_fpl_sum_2[cond & cond_2]
df_fpl_top3
```

```python
df_fpl_top3_summary = df_fpl_top3.groupby('team').sum()
df_fpl_top3_summary.reset_index(inplace=True)

cols = ['team','total_points']
df_fpl_top3_summary = df_fpl_top3_summary[cols].sort_values(by='total_points',ascending=False)
```

```python
df_fpl_top3_summary['fpl_rank'] = df_fpl_top3_summary.reset_index().index + 1
df_fpl_top3_summary
```

## Merge Data

```python
df_combined_top3 = df_epl.merge(df_fpl_top3_summary)

df_combined_top3['league_fpl_diff'] = df_combined_top3['league_rank'] - df_combined_top3['fpl_rank']

cols = ['team','league_points','league_rank','total_points','fpl_rank','league_fpl_diff']
df_combined_top3[cols]
```

## Total Points against Start Price

```python
df_fpl_min = df_fpl.groupby('full').min()
df_fpl_min.reset_index(inplace=True)

cols = ['full','round']
df_fpl_start_round = df_fpl_min[cols]
df_fpl_start_round.head()
```

```python
df_fpl_1st_gw = df_fpl.merge(df_fpl_start_round,on=cols)
df_fpl_1st_gw
```

```python
cols = ['full','value']
df_fpl_startprice = df_fpl_1st_gw[cols]
df_fpl_startprice.sort_values(by='value',ascending=False)
```

```python
df_fpl_totalpts_startprice = df_fpl_totalpts.merge(df_fpl_startprice,on='full')

cols = ['value','total_points']
order = [False,False]

df_fpl_totalpts_startprice.sort_values(by=cols,ascending=order)
```

## Charts

```python
plt.figure(figsize=(16,8))

x = df_fpl_totalpts_startprice['value']
y = df_fpl_totalpts_startprice['total_points']

sns.scatterplot(x=x,y=y)
```

```python
plt.figure(figsize=(16,8))

x = df_fpl_totalpts_startprice['value']
y = df_fpl_totalpts_startprice['total_points']

sns.boxplot(x=x,y=y)
```

```python
df_fpl_totalpts_adjuststartprice = df_fpl_totalpts_startprice.copy()

df_fpl_totalpts_adjuststartprice['value'].clip(upper=11.0,inplace=True)
df_fpl_totalpts_adjuststartprice['value'] = df_fpl_totalpts_adjuststartprice['value'].apply(np.floor)

df_fpl_totalpts_adjuststartprice.sort_values(by='value',ascending=False)
```

```python
plt.figure(figsize=(16,8))

x = df_fpl_totalpts_adjuststartprice['value']
y = df_fpl_totalpts_adjuststartprice['total_points']

sns.boxplot(x=x,y=y)
```
